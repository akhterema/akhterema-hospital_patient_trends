```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv("/content/hospital_patient_trends_dataset.csv")
print(data.head())
```

```python
#Basic info
print(data.info())
```

```python
#Cheek for missing value
print(data.isnull().sum())
```

```python
#Summary statistics
print(data.describe())
```

```python
#unique value in categorical colomns
print("Unique Categories:",data['Category'].unique())
print("Unique Labels:",data['Label'].unique())
```

```python
data = data.dropna()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data = data.drop_duplicates()
```

```python
print(data.dtypes)
```

```python
total_Patients=data.groupby('Category')['Number of Patients'].sum()
print(total_Patients)
```

```python
average_patients=data.groupby('Label')['Number of Patients'].mean()
print(total_Patients)
```

```python
max_days=data.groupby('Label')['Number of Patients'].max()
print(max_patients)
```

```python
data['Month']=data['Date'].dt.month
monthly_trends=data.groupby('Month')['Number of Patients'].sum()
print(monthly_trends)
```

```python
#Visualization
plt.figure(figsize=(8, 6))
plt.pie(total_Patients.values, labels=total_Patients.index, autopct='%1.1f%%', startangle=90)
plt.title("Total Patients by Category",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(average_patients.index, average_patients.values,color='blue')
plt.title('Average Patients by Category',fontsize=14)
plt.xlabel("Cetagory", fontsize=12)
plt.ylabel('Average_patients', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(monthly_trends.index, monthly_trends.values, marker='o', color='red')
plt.title("Monthly Trends by Category", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Patients", fontsize=12)
plt.show()
```
